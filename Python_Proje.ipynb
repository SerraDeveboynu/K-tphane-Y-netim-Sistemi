```python
class Kitap:
      """Kitap sınıfı: Kitapların kimliği,adı ve miktar bilgilerini tutar. """
def __init__(self,kitap_ID,kitap_ismi,miktar):
    self.kitap_ID=kitap_ID
    self.kitap_ismi=kitap_ismi
    self.miktar=miktar

    def __str__(self):
      return f"ID: {self.kitap_ID}, İsim: {self.kitap_ismi}, Miktar: {self.miktar}"

class Kullanıcı:
  """ Kullanıcı sınıfı: Kullanıcı bilgilerini ve ödünç aldığı kitapları tutar. """
def __init__(self,kullanıcı_ID,isim,ödünç_aldığı_kitaplar):
    self.kullanıcı_ID=kullanıcı_ID
    self.isim=isim
    self.ödünç_aldığı_kitaplar=[]

    def __str__(self):
      return f"ID: {self.kullanıcı_ID}, Kullanıcı isim: {self.isim}, Ödünç Aldığı Kitaplar: {len(self.ödünç_aldığı_kitaplar)}"

class Kütüphane:
  """ Kütüphane sınıfı: Kitap ve kullanıcı işlemlerini gerçekleştirir."""
def __init__(self):
    self.kitaplar = []
    self.kullanıcılar = []

    # .....Kitap İşlemleri......

    def kitap_ekle(self):
        kitap_ID= input("Kitap kimliğini girin:").strip()
        kitap_ismi = input("Kitap adını girin:").strip()
        try:
            miktar = int(input("Kitap miktarını girin:").strip())
        except ValueError:
            print("Geçerli bir miktar girin!")
            return

            for kitap in self.kitaplar:
              if kitap.kitap_ID == kitap_ID:
                print("Bu kimlikte bir kitap zaten mevcut! Miktar güncelleniyor.")
                kitap.miktar += miktar
                return

        yeni_kitap = Kitap(kitap_ID,kitap_ismi,miktar)
        self.kitaplar.append(yeni_kitap)
        print(f"{kitap_ismi} başarıyla eklendi!")

def kitaplari_listele(self):
    if not self.kitaplar:
      print("Kitap bulunmuyor!")
    else:
      print("\nKütüphanedeki kitaplar:")
      for kitap in self.kitaplar:
        print(kitap)

    def kitap_bul(self,kitap_ismi):
      for kitap in self.kitaplar:
        if kitap.kitap_ismi == kitap_ismi:
          return kitap
      return None

      # .....Kullanıcı İşlemleri......
      def kullanıcı_ekle(self):
        kullanıcı_ID= input("Kullanıcı kimliğini girin:").strip()
        kitap_ismi = input("Kitap adını girin:").strip()

        for kullanıcı in self.kullanıcılar:
          if kullanıcı.kullanıcı_ismi == kullanıcı_ID:
            print("Bu kimlikte bir kullanıcı zaten var!")
            return

        yeni_kullanıcı = Kullanıcı(kullanıcı_ID,kullanıcı_ismi)
        self.kullanıcılar.append(yeni_kullanıcı)
        print(f"Kullanıcı {kullanıcı_adı} başarıyla eklendi!")

        def kullanıcıları_listele(self):
          if not self.kullanıcılar:
            print("Kullanıcı bulunmuyor!")
          else:
            print("\nKütüphanedeki kullanıcılar:")
            for kullanıcı in self.kullanıcılar:
              print(kullanıcı)

      def kullanıcı_bul(self,kullanıcı_ID):
        for kullanıcı in self.kullanıcılar:
          if kullanıcı.kullanıcı_ID == kullanıcı_ID:
            return kullanıcı
        return None

        # ...... Ödünç ve İade İşlemleri

        def kitap_ödünç_al(self):
          kitap_ID = input("Ödünç alınacak kitabın kimliğini girin:").strip()
          kullanıcı_ID = input("Kullanıcı kimliğini girin:").strip()

          kitap = self.kitap_bul(kitap_ID)
          kullanıcı = self.kullanıcı_bul(kullanıcı_ID)

          if not kitap:
            print("Böyle bir kitap bulunmuyor!")
            return
          if not kullanıcı:
            print("Böyle bir kullanıcı bulunmuyor!")
            return
          if kitap.miktar <= 0:
            print(f"{kitap.ismi} kitabı stokta yok!")
            return

            kitap.miktar -= 1

            kullanıcı.ödünç_aldığı_kitaplar.append(kitap)
            print(f"{kitap.ismi} başarıyla {kullanıcı.ismi} kullanıcısına başarıyla ödünç verildi!")

            def kitap_iade_et(self):
              kitap_ID = input("İade edilecek kitabın kimliğini girin:").strip()
              kullanıcı_ID = input("Kullanıcı kimliğini girin:").strip()

              kitap = self.kitap_bul(kitap_ID)
              kullanıcı = self.kullanıcı_bul(kullanıcı_ID)

              if not kitap:
                print("Böyle bir kitap bulunmuyor!")
                return

              if not kullanıcı:
                  print("Böyle bir kullanıcı bulunmuyor!")
                  return

              for ödünç_kitap in kullanıcı.ödünç_aldığı_kitaplar:
                if ödünç_kitap.kitap_ID == kitap_ID:

                  kullanıcı.ödünç_aldığı_kitaplar.remove(ödünç_kitap)

                  kitap.miktar += 1
                  print(f"{kitap.ismi} başarıyla {kullanıcı.ismi} kullanıcısına başarıyla iade edildi!")
                  return

                  print(f"{kullanıcı.ismi} kullanıcısı bu kitabı ödünç almamış!")

          def kullanıcının_ödünç_aldığı_kitapları_listele(self):
            kullanıcı_ID = input("İade edilecek kitabın kimliğini girin:").strip()
            kullanıcı = self.kullanıcı_bul(kullanıcı_ID)

            if not kullanıcı:
              print("Böyle bir kullanıcı bulunmuyor!")
              return

            if not kullanıcı.ödünç_aldığı_kitaplar:
              print(f"{kullanıcı.ismi} kullanıcısının ödünç aldığı kitap bulunmuyor!")
            else:
              print(f"\n{kullanıcı.ismi} kullanıcısının ödünç aldığı kitaplar:")
              for kitap in kullanıcı.ödünç_aldığı_kitaplar:
                print(kitap)



    # Ana Program

    def main():
      kütüphane = Kütüphane()

      while True:
        print("\nKütüphane Sistemine Hoşgeldiniz...")
        print("1. Kitap Ekle")
        print("2.Kitapları Listele")
        print("3.Kullanıcı Ekle")
        print("4.Kullanıcıları Listele")
        print("5.Kitap Ödünç Al")
        print("6.Kitap İade Et")
        print("7.Kullanıcının Ödünç Aldığı Kitapları Listele")
        print("0.Çıkış")

        secim = input("Yapmak istediğiniz işlemi seçin (0-7):").strip()

        if secim == "1":
          kütüphane.kitap_ekle()
        elif secim == "2":
          kütüphane.kitapları_listele()
        elif secim == "3":
          kütüphane.kullanıcı_ekle()
        elif secim == "4":
          kütüphane.kullanıcıları_listele()
        elif secim == "5":
          kütüphane.kitap_ödünç_al()
        elif secim == "6":
          kütüphane.kitap_iade_et()
        elif secim == "7":

         kütüphane.kullanıcının_ödünç_aldığı_kitapları_listele()
        elif secim == "0":
          print("Sistemeden çıkılıyor...")
          break

        else:
          print("Geçersiz bir seçim yaptınız. Lütfen tekrar deneyin!")

if __name__ == "__main__":
    main()
```
